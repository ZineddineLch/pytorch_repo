```python
import torch
from torch import nn
import matplotlib.pyplot as plt
```

```python
## turn data into numbers
## pick or build a mode to learn representations as much as possible
```

```python
#no data we have to creat one data
weight=0.7
biase=0.3
data=torch.arange(0,1,0.02).unsqueeze(dim=1)
print(data)
y= weight * data + biase #model
print(y[:10])
```

```python
#split data
#into traininf data 60 80
#int validation
#into testing
training_data=int(0.8*len(data))
X_train,y_train=data[:training_data],y[:training_data]
X_test,y_test=data[training_data:],y[training_data:]
print(X_train,y_train)
print(X_test,y_test)
```

```python
import matplotlib.pyplot as plt

def plot_plot_prediction(X_train, y_train, X_test, y_test, predictions=None):
    plt.figure(figsize=(10,7))
    plt.scatter(X_train, y_train, c="b", s=4, label="training data")
    plt.scatter(X_test, y_test, c="g", s=4, label="testing data")

    if predictions is not None:
        plt.scatter(X_test, predictions, c="r", s=5, label="predicted data")

    plt.legend(prop={"size": 14})
    plt.xlabel("X values")
    plt.ylabel("Y values")
    plt.title("Training vs Testing vs Predictions")
    plt.show()
```

```python
#build a model
#create linear regression model
from torch import nn
class linear_regression_model(nn.Module):
  def __init__(self):
    super().__init__()
    self.weight=nn.Parameter(torch.randn(1,requires_grad=True))
    self.biase=nn.Parameter(torch.randn(1,requires_grad=True));
  def forward(self,x:torch.tensor):
      return self.weight*x+self.biase;
```

```python
##Â© start with random value w and b
#look at the trainging data and adjust the random values to better represend the ideal values (weight and biase)
''' we call this
1-gradient desecne
2-backpropagation
'''
```

```python
import torch
torch.manual_seed(42)
model_0=linear_regression_model()
list(model_0.parameters())
```

```python

#making prediction using the torch.inference_mode()
with torch.inference_mode():
  y_pred=model_0(X_test)
y_pred
```

```python
y_pred,y_test
```

```python
plot_plot_prediction(X_train, y_train, X_test, y_test, predictions=y_pred)
```

```python
#training aÃ´ur idea means findting a new known parameters form unknown paprmaters
```

```python
#setting up the loss function and make an optimazation
loss_function=nn.L1Loss()
loss_function
```

```python
optimizer=torch.optim.SGD(params=model_0.parameters(),lr=0.01, momentum=0, dampening=0, weight_decay=0, nesterov=False, maximize=False, foreach=None, differentiable=False, fused=None)
optimizer
```

```python
##loop through the data
## forward pass it mean make the predection
#calculate the loss and make a optimization also called the backward propgation
```

```python
epochs=150
for epoch in range(epochs):
  model_0.train()
  y_pred=model_0(X_train);
  loss=loss_function(y_pred,y_train)
  optimizer.zero_grad()
  loss.backward()

  optimizer.step()

  model_0.eval()
  with torch.inference_mode():
    y_pred_test=model_0(X_test)
    test_loss=loss_function(y_pred_test,y_test)
    print(f"test loss:{test_loss}")


```

```python
#testing steps
list(model_0.parameters())
```

```python
plot_plot_prediction(X_train, y_train, X_test, y_test, predictions=y_pred_test)
```
